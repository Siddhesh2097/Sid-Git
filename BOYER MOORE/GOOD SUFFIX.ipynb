```python
def Good_suffix(T,P):
    end=len(P)-1
    endp=len(P)-1
    match=0
    count=0
    t=[]
    startp=0
    shift=0
    endt=end
    while count<len(T):
        print("",T)
        print(" "*startp,P)
        if T[end]==P[endp]:
            t.append(T[end])
            end-=1
            endp-=1
            match+=1    
            if match==len(P):
                print("\nComplete match found at {} index \n".format(startp))
                shift=1
                endt+=shift
                startp+=shift
                end=endt
                endp=len(P)-1
                match=0
                count=endt
                t=[]

        elif T[end]!=P[endp] and match==0: #HERE BC IS REMOVED AND ONLY GS IS KEPT
            shift=len(P)
            endt+=shift
            startp+=shift
            end=endt
            endp=len(P)-1
            match=0
            count=endt
            t=[]
        elif T[end]!=P[endp] and match>0:
            s=''
            for ele in range(len(t)-1,-1,-1):
                s+=t[ele]
            t=[]
            if s in P[0:endp+1] and match==1:###This is only for one match
                for char_index in range(endp,0,-1):
                    if P[char_index]==s:  #Only one character has been taken...............
                        break
                shift=(end+1)-(startp+char_index)
            elif s in P[0:endp+1] and match>1:  ##THis is if entire s matches.........
                p_index=P.index(s)
                shift=(end+1)-(startp+p_index)
            elif s not in P[0:endp]:  ##If complete s is not there but some sub is found
                s_sub=[]
                for i in range(len(s)-1): #To find substrings of s
                    for j in range(len(s)):
                        if s[i:j+1]==s:
                            pass
                        elif s[i:j+1]=='':
                            pass
                        elif s[i:j+1] in s_sub:
                            pass
                        else:
                            s_sub.append(s[i:j+1])
                pprefix=[]
                for i in range(len(P)):  #To find substrings specifically prefix of p
                    x=P[:i+1]
                    if x==s:
                        pass
                    elif x=='':
                        pass
                    elif x in pprefix:
                        pass
                    elif len(x)<=(len(s)-1):
                        pprefix.append(x)
                match_sub=[]
                for s1 in s_sub:    #To find matches of s_sub and p_sub
                    for ps in pprefix:
                        if s1==ps:
                            match_sub.append(s1)
                if len(match_sub)==1:
                    sub_found=max(match_sub)
                    p_prefix_index=P.index(sub_found)
                    s_suff_index=end+(len(s)-1-s[::-1].index(sub_found))+1
                elif len(match_sub)>1:
                    sub_found=max(match_sub)
                    p_prefix_index=P.index(sub_found) #THis is index of prefix in P which is always going to be zero
                    #We need to find index of suf_found in T
                    s_suff_index=end+s.index(sub_found)+1 #This is index of suffix in s
                else:
                    shift=endt-startp
                if len(match_sub)>=1:
                    if p_prefix_index==0:
                        shift=s_suff_index-(startp+p_prefix_index)
                    else:
                        shift=endt-startp

            endt+=shift
            startp+=shift
            end=endt
            endp=len(P)-1
            match=0
            count=endt
            t=[]
            
            
            
if __name__ == '__main__':
    T='CGTGCCTACTTACTTACTTACTTACGCGAA'
    P='CTTACTTAC'
    #T='GCTTCTGCTACCTTTTGCGCGCGCCTTTTGCCGCGGAACCTTTTGCATGCATGCATCCTTTTGCAGCTCGCGCGCCTTTTGCGCGCGCGCGCCTTTTGCAGGC'
    #P='CCTTTTGC'
    #T='GTTATAGCTGATCGCGGCGTAGCGGCGAA'
    #P='GTAGCGGCG'
    #T='GCTTCTGCTACCTTTTGCGCGCGCGGAACCTTTTGCGCGCGCCAACCTTTTGCAA'
    #P='CCTTTTGC'
    Good_suffix(T,P)
        
```
