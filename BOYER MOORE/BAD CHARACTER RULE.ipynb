# BAD CHARACTER RULE.............................................................................

```python
def bad_character_rule(T,P):
    ###Bad character rule
    endtp=len(P)-1  ###For shift
    end=len(P)-1
    startp=0
    endp=len(P)-1
    count=endtp #Count till end of the text .i.e endtp will be matching last character .....
    shift=0
    match=0
    while count<len(T):  ##count is till the length of T
        print("",T)
        print(" "*startp,P)
        if T[end]==P[endp]:
            end-=1
            match+=1
            endp-=1
            if match==len(P):
                print("Match has been found at {} index \n\n".format(startp))
                shift=1
                endtp+=shift
                startp+=shift
                count=endtp
                end=endtp
                endp=len(P)-1
                match=0
        elif T[end]!=P[endp]:
            if T[end] in P[0:endp]:
                for char_index in range(endp,0,-1):
                    if P[char_index]==T[end]:
                        break
                shift=end-(char_index+startp)
            else:
                var1=end+len(P)
                var2=var1-end
                var3=endtp-end
                shift=var2-var3
            endtp+=shift
            startp+=shift
            end=endtp
            endp=len(P)-1
            match=0
            count=endtp
    
    
if __name__ == '__main__':
    T='GCTTCTGCTACCTTTTGCGCGCGCCTTTTGCCGCGGAACCTTTTGCATGCATGCATCCTTTTGCAGCTCGCGCGCCTTTTGCGCGCGCGCGCCTTTTGCAGGC'
    P='CCTTTTGC'
    bad_character_rule(T,P)
    
```
