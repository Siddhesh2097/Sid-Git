```python
#S1='CGTGAATTCAT'
#S2='GACTTAC'
#S1='TGTTACGG'
#S2='GGTTGACTA'
#S1='GGCTCAATCA'
#S2='ACCTAAGG'
S1='TCAGTTGCC'
S2='AGGTTG'
#S1=input("Enter reference sequence : ")
#S2=input("Enter aligning sequence : ")

GAP='-'
S1=GAP*2+S1
S2=GAP*2+S2
mat_row=len(S2)
mat_col=len(S1)
match=int(input("Enter match score: "))
mismatch=int(input("Enter mismatch score: "))
gap=int(input("Enter gap penalty: "))

#Creating matrix of 0s........................................................
mat=[[0 for j in range(mat_col)] for i in range(mat_row)]

#Creating trace_matrix of 0s....................................................
trace_mat=[[0 for j in range(mat_col)] for i in range(mat_row)]

#Adding string S1 to first row..................................................
for i in range(1):
    for j in range(mat_col):
        mat[i][j]=S1[j]
        trace_mat[i][j]=S1[j] #..............................For trace_mat matrix

#Adding string S2 to first column................................................
for i in range(mat_row):
    for j in range(1):
        mat[i][j]=S2[i]
        trace_mat[i][j]=S2[i] #..............................For trace_mat matrix
        
#mat

##Start filling actual values...................................................
##fill value of mat[row][col]...................................................

for row in range(2,mat_row):
    for col in range(2,mat_col):
        values=[]
        
        value_up=mat[row-1][col]+gap #.......................Up adjacent
        values.append(value_up)
        
        value_left=mat[row][col-1]+gap #.....................Left adjacent
        values.append(value_left)
        #Calculating diagonal....................................................
        if mat[row][0]==mat[0][col]:
            value_diag=mat[row-1][col-1]+match
            values.append(value_diag)
        else:
            value_diag=mat[row-1][col-1]+mismatch
            values.append(value_diag)
            
        best_val=max(value_up,value_left,value_diag) #.......Max value of all three is given to that box
        
        #Operation for trace matrix.............................................
        if best_val==value_up and best_val==value_left: #....BEST==UP AND LEFT
            trace_mat[row][col]=['Up','Left']
        elif best_val==value_left and best_val==value_diag: #BEST==LEFT AND DIAG
            trace_mat[row][col]=['Left','Diag']
        elif best_val==value_up and best_val==value_diag: #..BEST==UP AND DIAG
            trace_mat[row][col]=['Up','Diag']
        elif best_val==value_up:    #........................BEST=UP
            trace_mat[row][col]='Up'
        elif best_val==value_left:  #........................BEST=LEFT
            trace_mat[row][col]="Left"
        elif best_val==value_diag:  #........................BEST=DIAG
            trace_mat[row][col]="Diag"
        
        #SMITH WATERMAN COMES ........................................................
        if best_val < 0:
            best_val=0
        
        mat[row][col]=best_val  ##...........................Adding best value to the box
        
#mat

#print(trace_mat)

#####CODE TO FIND THE MAXIMUM VALUE

maximum=0
for row in range(2,mat_row):
    for col in range(2,mat_col):
        if mat[row][col] >= maximum:
            maximum=mat[row][col]
#print(maximum , find_row , find_col)
        
    
##We have found the col and row

######################################################################################################
#trace_mat  #...................................................Checking trace_mat matrix
######################################################################################################

for cur_row in range(2,mat_row):
    for cur_col in range(2,mat_col):
        A1=''
        A2=''
        if mat[cur_row][cur_col]==maximum:
            ##Start tracing............................................................
            while trace_mat[cur_row][cur_col]=='Diag' or trace_mat[cur_row][cur_col]=='Left' or trace_mat[cur_row][cur_col]=='Up' or trace_mat[cur_row][cur_col]==['Up','Diag'] or trace_mat[cur_row][cur_col]==['Left','Diag']:
                if mat[cur_row][cur_col]==0:
                    break

                if trace_mat[cur_row][cur_col]=='Diag':
                    #print('{}() came from Diag'.format(cur_row,cur_col))
                    A1+=trace_mat[0][cur_col]
                    A2+=trace_mat[cur_row][0]
                    cur_row-=1
                    cur_col-=1 #...............................................Here we add gap to S2 sequence i.e to A2    
                elif trace_mat[cur_row][cur_col]=='Left':
                    #print("{}{} came from Left".format(cur_row,cur_col))
                    A1+=trace_mat[0][cur_col]
                    A2+=GAP  #.................................................Here we'll add gap
                    cur_col-=1    
                elif trace_mat[cur_row][cur_col]=="Up":
                    #print("{}{} came from Up".format(cur_row,cur_col))
                    A1+=GAP  #..................................................Here we'll add GAP
                    A2+=trace_mat[cur_row][0] 
                    cur_row-=1
                elif trace_mat[cur_row][cur_col]==['Up','Diag']:
                    up_value=mat[cur_row-1][cur_col]
                    diag_value=mat[cur_row-1][cur_col-1]
                    if up_value>diag_value:
                        #print('Up value is greater in up,diag')
                        A1+=GAP
                        A2+=trace_mat[cur_row][0]
                        cur_row-=1
                    elif diag_value>up_value:
                        #print('diag is greater in up,diag')
                        A1+=trace_mat[0][cur_col]
                        A2+=trace_mat[cur_row][0]
                        cur_row-=1
                        cur_col-=1
                elif trace_mat[cur_row][cur_col]==['Left','Diag']:
                    left_value=mat[cur_row][cur_col-1]
                    diag_value=mat[cur_row-1][cur_col-1]
                    if left_value>diag_value:
                        #print('left is greater in left,diag')
                        A1+=trace_mat[0][cur_col]
                        A2+=GAP
                        cur_col-=1
                    elif diag_value>left_value:
                        #print('diag is greater in left,diag')
                        A1+=trace_mat[0][cur_col]
                        A2+=trace_mat[cur_row][0]
                        cur_row-=1
                        cur_col-=1
            

            #Then we simply reverse it..........................................................
            A1=A1[::-1]
            A2=A2[::-1]

            print('Reference Sequence :',A1)
            print('Local Alignment    :',A2)
            
            ####################################################################################################
            #To calculate score............................................................
            score=0
            count=0
            while count<len(A1):
                for a1 in range(1):
                    for a2 in range(len(A2)):
                        if A1[a1+a2]==A2[a2]:
                            score+=match
                        elif A1[a1+a2]=='-' or A2[a2]=='-':
                            score+=gap
                        else:
                            score+=mismatch

                        count+=1
            print('The score of the alignment sequence {} and {} is {}'.format(A1,A2,score))

```
