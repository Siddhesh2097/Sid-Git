```python
##BWT...................................................................................
def burrows_wheeler_transform(string):  ##THIS GIVES US BWT
    EOF="$"
    string+=EOF
    #string=SOF+string
    ready_str=string
    count=0
    cyclic_list=[string]
    while count<len(string)-1:

        ope_str=ready_str
        ready_str=""
        extract=ope_str[len(ope_str)-1]
        ready_str+=extract

        for char_in in range(len(ope_str)-1):
            ready_str+=ope_str[char_in]
        cyclic_list.append(ready_str)
        
        count+=1
    sorted_list=sorted(cyclic_list)
    bwt=""
    for ele in sorted_list:
        bwt+=ele[len(ele)-1]
        
    return (bwt)

#PATTERN MATCHING FUNCTION........................................................................
def pattern_match(bwt,pattern,string):
    ##CODE for last column........................................................................
    count_A=0
    count_T=0
    count_G=0
    count_C=0

    last_col=[]
    for ele in bwt:
        if ele=='$':
            last_col.append([ele])
        elif ele=='A':
            if ['A',0] in last_col:
                count_A+=1
            last_col.append([ele,count_A])
        elif ele=='T':
            if ['T',0] in last_col:
                count_T+=1
            last_col.append([ele,count_T])
        elif ele=='G':
            if ['G',0] in last_col:
                count_G+=1
            last_col.append([ele,count_G])
        elif ele=='C':
            if ['C',0] in last_col:
                count_C+=1
            last_col.append([ele,count_C])


    ##WE HAVE OUR LAST COLUMN READY......................................

    ##Counting A,T,G,C repeats................................................................

    count_A+=1
    count_C+=1
    count_G+=1
    count_T+=1

    ####Making first column string..............................................................
    first_col=''
    first_col='$'+count_A*'A'+count_C*'C'+count_G*'G'+count_T*'T'

    #CODE for first column.................................................................
    count_A1=0
    count_T1=0
    count_G1=0
    count_C1=0

    first_column=[]
    for ele in first_col:
        if ele=='$':
            first_column.append(ele)
        elif ele=='A':
            if ['A',0] in first_column:
                count_A1+=1
            first_column.append([ele,count_A1])
        elif ele=='T':
            if ['T',0] in first_column:
                count_T1+=1
            first_column.append([ele,count_T1])
        elif ele=='G':
            if ['G',0] in first_column:
                count_G1+=1
            first_column.append([ele,count_G1])
        elif ele=='C':
            if ['C',0] in first_column:
                count_C1+=1
            first_column.append([ele,count_C1])
    
    ##CODE FOR SUFFIX TREE............................................................
    
    EOF='$'
    S=string+EOF
    suf_arr=[]
    for i in range(len(S)):
        suf_arr.append([i,S[i:]])

    suf_arr.sort(key=lambda x:x[1])

    


    #####################################...........................................
    ##PATTERN MATCHING CODE........................................................
    count_pat=len(pattern)

    ele=pattern[count_pat-1] 

    if ele=='A':
        loop_run=count_A
    elif ele=='T':
        loop_run=count_T
    elif ele=='G':
        loop_run=count_G
    elif ele=='C':
        loop_run=count_C

    match_str=''
    
    for i in range(loop_run):
        match_str=''
        count_pat=len(pattern)  ##count_pat=2
        ele=pattern[count_pat-1]
        while count_pat>0:
            match_str=last_col[last_col.index([ele,i])][0]+match_str
            find_first=first_column.index([ele,i])
            find_last_ele=last_col[find_first]  ##here we have ['G',0]
            if find_last_ele[0]==pattern[count_pat-2]: 
                count_pat-=1
                ele=find_last_ele[0]
                i=find_last_ele[1]
            elif find_last_ele[0]!=pattern[count_pat-2]:
                count_pat=0  ##again we go to start of loop

        if match_str==pattern:
                match_at_index=suf_arr[find_first][0]
                print("We have found the match of the pattern {} in the string {} at index {}\n".format(pattern,S,match_at_index))
                match_str=''
                
                
#####CALLING FUNCTIONS...................................................................................


S='AGCTTCGACGATCGACGACCGATCGACCGAT'
string=S
bwt=burrows_wheeler_transform(string)
pattern='GAC'
print("The Burrows Wheeler Transform of {} is {}".format(string,bwt))
pattern_match(bwt,pattern,string)
```
