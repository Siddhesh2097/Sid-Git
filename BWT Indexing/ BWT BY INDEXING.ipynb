# BWT WITH INDEXING

```python
##BWT...................................................................................
def burrows_wheeler_transform(string):  ##THIS GIVES US BWT
    EOF="$"
    string+=EOF
    #string=SOF+string
    ready_str=string
    count=0
    cyclic_list=[string]
    while count<len(string)-1:

        ope_str=ready_str
        ready_str=""
        extract=ope_str[len(ope_str)-1]
        ready_str+=extract

        for char_in in range(len(ope_str)-1):
            ready_str+=ope_str[char_in]
        cyclic_list.append(ready_str)
        
        count+=1
    sorted_list=sorted(cyclic_list)
    bwt=""
    for ele in sorted_list:
        bwt+=ele[len(ele)-1]
        
    return (bwt)

 #THIS IS BASIC INVERSE FUNCTION.....................................................................
def inverse_bur_whe_tran(bwt):
    EOF="$"
    ope_str=bwt
    count=0
    
    while count<len(bwt):
        str_sort=sorted(ope_str)
        ope_str=[]
        for char in range(1):
            for ele in range(len(str_sort)):
                x=bwt[char+ele]+str_sort[ele]
                ope_str.append(x)
        count+=1
    ##take out string from str_sort with ending EOF
    for ele in str_sort:
        if ele[len(ele)-1]==EOF:
            return (ele)



########INV WITH INDEX FUNCTION..........................................................................
def inv_bwt_index(bwt):
    ##CODE for last column........................................................................
    count_A=0
    count_T=0
    count_G=0
    count_C=0

    last_col=[]
    for ele in bwt:
        if ele=='$':
            last_col.append([ele])
        elif ele=='A':
            if ['A',0] in last_col:
                count_A+=1
            last_col.append([ele,count_A])
        elif ele=='T':
            if ['T',0] in last_col:
                count_T+=1
            last_col.append([ele,count_T])
        elif ele=='G':
            if ['G',0] in last_col:
                count_G+=1
            last_col.append([ele,count_G])
        elif ele=='C':
            if ['C',0] in last_col:
                count_C+=1
            last_col.append([ele,count_C])


    ##WE HAVE OUR LAST COLUMN READY......................................

    ##Counting A,T,G,C repeats................................................................

    count_A+=1
    count_C+=1
    count_G+=1
    count_T+=1
    print('count of a',count_A)
    print('count of t',count_T)
    print('count of g',count_G)
    print('count of c',count_C)


    ####Making first column string..............................................................
    first_col=''
    first_col='$'+count_A*'A'+count_C*'C'+count_G*'G'+count_T*'T'

    #CODE for first column.................................................................
    count_A1=0
    count_T1=0
    count_G1=0
    count_C1=0

    first_column=[]
    for ele in first_col:
        if ele=='$':
            first_column.append(ele)
        elif ele=='A':
            if ['A',0] in first_column:
                count_A1+=1
            first_column.append([ele,count_A1])
        elif ele=='T':
            if ['T',0] in first_column:
                count_T1+=1
            first_column.append([ele,count_T1])
        elif ele=='G':
            if ['G',0] in first_column:
                count_G1+=1
            first_column.append([ele,count_G1])
        elif ele=='C':
            if ['C',0] in first_column:
                count_C1+=1
            first_column.append([ele,count_C1])

    ##Decompressing............................................................................    
    count=0
    trace_back=0
    decom_str=''

    #We start by adding EOF to decompre_str.....................................................
    decom_str+=first_column[trace_back]
    decom_str

    while count<len(bwt)-1:
        adj=last_col[trace_back]
        decom_str=adj[0]+decom_str 
        trace_back=first_column.index(adj)
        count+=1

    return (decom_str)

#####CALLING FUNCTIONS...................................................................................


S='AGATCAGA'
string=S
bwt=burrows_wheeler_transform(string)
inv_bwt=inverse_bur_whe_tran(bwt)
inv_bwt_index=inv_bwt_index(bwt)
print("The Burrows Wheeler Transform of {} is {}".format(string,bwt))
print("The Inverse Burrows Wheeler Transform of {} is {}".format(bwt,inv_bwt))
print("The Inverse Burrows Wheeler Transform of {} with indexing method is {}".format(bwt,inv_bwt_index))



```
