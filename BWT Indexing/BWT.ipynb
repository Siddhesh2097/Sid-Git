```python
def suffix_array(S):
    EOF='$'
    S=S+EOF
    suf_arr=[]
    for i in range(len(S)):
        suf_arr.append([i,S[i:]])

    suf_arr.sort(key=lambda x:x[1])

    return (suf_arr)

S='AGATCAGA'

suf_arr=suffix_array(S)
```
